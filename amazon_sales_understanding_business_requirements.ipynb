```python
# For the 2 missing 'rating_count' values, we will fill them with 0.
# This assumes no ratings were given for those products.
df['rating_count'] = df['rating_count'].fillna('0')
```

```python
import re # Import regular expression module for text cleaning
```

```python
def clean_currency_and_convert(text):
    """
    Cleans strings like '₹1,299' or '1,299' and converts to float.
    """
    if isinstance(text, str):
        # Remove currency symbols (₹, $, etc.), commas, and any extra spaces
        text = re.sub(r'[₹$,]', '', text).strip()
    try:
        return float(text)
    except:
        # If conversion fails (e.g., for empty strings), return NaN
        return np.nan
```

```python
def clean_percentage(text):
    """
    Cleans strings like '46%' and converts to float (e.g., 46.0).
    """
    if isinstance(text, str):
        text = text.replace('%', '').strip()
    try:
        return float(text)
    except:
        return np.nan
```

```python
# 1. Clean and convert price columns to numeric (float)
print("Cleaning price columns...")
df['discounted_price'] = df['discounted_price'].apply(clean_currency_and_convert)
df['actual_price'] = df['actual_price'].apply(clean_currency_and_convert)
```

```python
# 2. Clean and convert discount_percentage to numeric (float)
print("Cleaning discount column...")
df['discount_percentage'] = df['discount_percentage'].apply(clean_percentage)
```

```python
# 3. Clean and convert rating to numeric (float). Some might have text like '|'.
print("Cleaning rating column...")
df['rating'] = pd.to_numeric(df['rating'], errors='coerce') # 'coerce' turns errors to NaN
```

```python
# 4. Clean rating_count: remove commas and convert to integer.
print("Cleaning rating_count column...")
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '', regex=False)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce').fillna(0).astype(int)
```

```python
# 5. (Optional but recommended) Clean the 'category' column.
# It has long strings with pipes '|'. Let's extract the main category (first one).
print("Extracting main category...")
df['main_category'] = df['category'].astype(str).apply(lambda x: x.split('|')[0].strip())
```

```python
# === TASK: Identify and Handle Outliers (for numerical columns) ===
# Let's check the basic statistics to spot obvious outliers in prices and ratings.
print("--- Data Summary for Outlier Check ---")
print(df[['discounted_price', 'actual_price', 'discount_percentage', 'rating']].describe())
```

```python
# Based on the .describe() output, you might decide to cap outliers.
# For example, if discount_percentage is >100% (which is unrealistic), cap it.
print("Checking for unrealistic discounts (>100%)...")
unrealistic_discounts = df[df['discount_percentage'] > 100]
print(f"Found {len(unrealistic_discounts)} entries with discount > 100%.")
# Cap them at 100
df.loc[df['discount_percentage'] > 100, 'discount_percentage'] = 100.0
```

```python
# === TASK: Create New Columns for Analysis ===
print("Creating new analysis columns...")
# 1. Calculate the absolute discount amount (actual - discounted)
df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

```python
# 2. Create a flag for high-impact products (e.g., high rating & many reviews)
# Let's define "high-impact" as rating >= 4.0 AND rating_count above the 75th percentile.
rating_count_threshold = df['rating_count'].quantile(0.75)
df['is_high_impact'] = (df['rating'] >= 4.0) & (df['rating_count'] > rating_count_threshold)
```

```python
# 3. Create a simple value-for-money score (Discount % * Rating)
# Normalize rating to a 0-1 scale first (divide by 5).
df['value_score'] = (df['discount_percentage'] / 100) * (df['rating'] / 5)
```

```python
# === TASK: Final Check and Save ===
print("=== Final Data Info After Cleaning ===")
print(df.info())
```
